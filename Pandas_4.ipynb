```
data = pd.read_csv('iris.csv')
data.head(10)
```

```
for i in data.index:
  if data.loc[i, 'sepal_length'] >=8.0:
    data.loc[i, 'sepal_length'] = 7.0
print(data.to_string())  
```

```
print(data['sepal_length'])
```

```
data.loc[0, 'sepal_length'] = np.nan
data
```

```
for i in data.index:
  if pd.isnull(data.loc[i,'sepal_length']) == True:
    data.drop(data.index[i], inplace=True) 
data
```

```
diab = pd.read_csv('diabetes.csv')
diab.columns=["Pregnancies", "Glucose", "BloodPressure", "SkinThickness(mm)", "Insulin(mu U/ml)", "BMI", "DiabetesPedigreeFunction", "Age(Years)", "Outcome"]
print(diab)
```

```
dib = diab.loc[0:50,['Glucose', 'BloodPressure','BMI', 'Age(Years)']]
print(dib)
```

```
dib.duplicated()
```

```
dib.dropna(inplace=True)
dib
```

```
dup = dib.duplicated()
for i in dup.index:
  if dup.loc[i]==True:
    print(i)
```
